### 3. Melihat Outliers

#### A. Boxplot

```python
plt.figure(figsize=(10,10))
posisi = 1
for setiap_kolom in saas_numerik.columns:
    plt.subplot(3,2,posisi)
    sns.boxplot(x=saas_numerik[setiap_kolom])
    posisi += 1
    plt.tight_layout()
```

#### B. Interquartil Range (IQR)

```python
persentage = len(saas_numerik)

jumlah_outlier = []
for setiap_kolom in saas_numerik.columns:
    q1 = saas_numerik[setiap_kolom].describe()['25%']
    q3 = saas_numerik[setiap_kolom].describe()['75%']
    iqr = q3 - q1
    wish = iqr * 1.5
    upper_bond = q3 + wish
    lower_bond = q1 - wish

    outliers = len(saas_numerik[(saas_numerik[setiap_kolom]<lower_bond)|(saas_numerik[setiap_kolom]>upper_bond)])
    jumlah_outlier.append(outliers)

outliers_df = pd.DataFrame({
                            'Kolom':saas_numerik.columns,
                            'Jumlah Outliers':jumlah_outlier,
                            })

outliers_df['Percentage %'] = round((outliers_df['Jumlah Outliers']/persentage)*100,2)
outliers_df
```

#### C. ZScore

```python
persentage = len(saas_numerik)

jumlah_outlier = []
for setiap_kolom in saas_numerik.columns:
    mean = saas_numerik[setiap_kolom].mean()
    std = saas_numerik[setiap_kolom].std()
    zscore = abs((saas_numerik[setiap_kolom]-mean)/std)

    outliers = len(zscore[zscore>2])
    jumlah_outlier.append(outliers)

zscore_df = pd.DataFrame({
                          'Kolom':saas_numerik.columns,
                          'Jumlah Outliers':jumlah_outlier
                        })

zscore_df['Percentage %'] = round((zscore_df['Jumlah Outliers']/persentage)*100,2)
zscore_df
```

Uji Hipotesis untuk Menganalisis Perbedaan Profit Antar Negara
-   1. Hipotesis Nol (H0):
    -   H0 : Tidak ada perbedaan signifikan antara profit dengan kerugian pada seluruh negara
-   2. Hipotesis Alternatif (Ha):
    -   Ha : Terdapat perbedaan signifikan antara profit dengan kerugian pada seluruh negara

Metode Uji Statistik
-   Untuk menguji hipotesis di atas, akan digunakan Uji Mann-Whitney U. Uji ini dipilih karena data yang yang di uji tidak berdistribusi normal (non-parametric) dan uji ini memiliki kemampuannya untuk membandingkan dua kelompok independen yang tidak memerlukan asumsi distribusi normal, serta dapat menangani ketidak seimbangan ukuran sampel antara kelompok.

Variabel yang Digunakan:
-   X: Distribusi kerugian di negara-negara yang mengalami kerugian.
-   Y: Distribusi profit di negara-negara yang menghasilkan profit.

Scipy Documentation on mannwhitneyu Function:
Dokumentasi resmi dari library SciPy di Python, yang menyediakan implementasi uji Mann-Whitney U. Di sini dijelaskan bahwa fungsi mannwhitneyu dapat digunakan dengan kelompok yang memiliki ukuran yang berbeda.

```python
list_rugi = []
for i in saas_country['Profit']:
    if i < 0.0:
        ket = 'Net Loss'
        list_rugi.append(ket)
    else:
        ket = 'Profit'
        list_rugi.append(ket)

df_uji = pd.DataFrame({'Keuntungan':list_rugi,'Total':saas_country['Profit']})

from scipy.stats import mannwhitneyu

# cek dengan uji statistika

H0 = 'Tidak ada perbedaan signifikan antara profit dengan kerugian pada seluruh negara'
Ha = 'Terdapat perbedaan signifikan antara profit dengan kerugian pada seluruh negara'
stat, pval = mannwhitneyu(df_uji[df_uji['Keuntungan']=='Profit']['Total'],
                          df_uji[df_uji['Keuntungan']=='Net Loss']['Total'])
if pval > 0.05:
    print(f"H0 diterima: {H0} (pvalue = {pval})")
else:
    print(f"H0 ditolak: {Ha} (pvalue = {pval})")
```

-   Berdasarkan hasil uji hipotesis dengan menggunakan Uji Mann-Whitney U, berikut adalah kesimpulan dari analisis yang dilakukan

-   Hasil Uji:
    -   Kesimpulan:
        -   Dengan p-value yang sangat rendah P-Value: 1.5038824724344121e-06 (atau 0.0000015), ada cukup bukti untuk menolak hipotesis H0 dan menerima Ha. Ini menunjukkan bahwa terdapat perbedaan signifikan antara distribusi profit di negara-negara yang memperoleh profit positif dan negara-negara yang mengalami kerugian.

    -   Interpretasi:
        -   Ada Perbedaan Signifikan: Hasil ini mengindikasikan bahwa distribusi di negara-negara yang memperoleh profit berbeda secara signifikan dari distribusi di negara-negara yang mengalami kerugian.

    -   Implikasi: 
        -   Perbedaan signifikan ini dapat menunjukkan bahwa faktor-faktor yang mempengaruhi profitabilitas di negara-negara yang memperoleh profit sangat berbeda dari faktor-faktor yang menyebabkan kerugian di negara-negara lainnya. Hal ini mungkin memerlukan analisis lebih lanjut untuk memahami penyebab perbedaan tersebut dan untuk merumuskan strategi yang lebih baik dalam meningkatkan profitabilitas dan mengurangi kerugian.

#### 2. Identifikasi Hidden Cost

Pemahaman yang mendalam tentang biaya merupakan elemen kunci untuk keberhasilan bisnis. Namun, tidak semua biaya terlihat jelas pada data yang dimiliki, Hidden costs atau biaya tersembunyi adalah jenis pengeluaran yang sering kali tidak terdeteksi atau terabaikan dalam data yang ada, tetapi dapat secara signifikan mempengaruhi profitabilitas suatu. engidentifikasi dan mengelola hidden costs sangat penting bagi bisnis untuk menghindari kebocoran keuangan dan memastikan bahwa profit yang diharapkan benar-benar tercapai.

-	Profit biasanya dihitung setelah dikurangi semua biaya yang terkait dengan penjualan produk atau layanan tersebut. Biaya ini bisa mencakup:
    -	Biaya operasional: Seperti biaya server, pemeliharaan, dukungan pelanggan, dan lisensi pihak ketiga yang dibutuhkan untuk menjalankan layanan.
    -	Biaya langsung lainnya: Misalnya, pajak import, biaya distribusi, atau pengeluaran lain yang terkait dengan produk atau layanan.
    -   Kolom hidden cost ini dibuat untuk mencari tahu seberapa besar biaya yang tidak tercantum pada data, sehingga kita mengetahui total selisih antara profit   dengan total sales.


- Membuat kolom Hidden Cost

```python
# Membuat kolom baru yaitu Gross Profit, agar mengetahui profit bersih setelah discount
gross_profit = saas_sales['Sales']*saas_sales['Quantity']
discount = gross_profit * saas_sales['Discount']
saas_sales['Gross Profit'] = gross_profit-discount # Gross Profit adalah Quantity * Sales dan dikurangi dengan Discount

# Membuat kolom hidden cost
hiddencost = []  
for profit, gross_profit in zip(saas_sales['Profit'], saas_sales['Gross Profit']):
    if profit < 0.0:
        isi = profit + (-1*gross_profit)
    else:
        isi2 = profit - gross_profit
        isi = isi2  
    hiddencost.append(isi)

hidden_cost = pd.DataFrame({'Hidden_Cost':hiddencost})
saas_sales['Hidden Cost'] = hidden_cost
saas_sales.head()
```

Uji Hipotesis untuk Menganalisis Perbedaan Profit Antar Negara
-   1. Hipotesis Nol (H0):
    -   H0 : Tidak ada perbedaan signifikan antara total pajak import pada negara profit dengan total pajak import pada negara rugi
-   2. Hipotesis Alternatif (Ha):
    -   Ha : Terdapat perbedaan signifikan antara total pajak import pada negara profit dengan total pajak import pada negara rugi

Metode Uji Statistik
-   Untuk menguji hipotesis di atas, akan digunakan Uji Mann-Whitneyu Two Tail. Uji ini dipilih karena data yang yang di uji tidak berdistribusi normal (non-parametric) dan uji ini memiliki kemampuannya untuk membandingkan dua kelompok independen yang tidak memerlukan asumsi distribusi normal, serta dapat menangani ketidak seimbangan ukuran sampel antara kelompok.

Variabel yang Digunakan:
-   X: Total Import pada setiap negara.
-   Y: Profit pada setiap negara.

```python
list_tax = []
for i in tax_rate_transaction_num['Profit']:
    if i < 0.0:
        ket = 'Net Loss'
        list_tax.append(ket)
    else:
        ket = 'Profit'
        list_tax.append(ket)

df_uji = pd.DataFrame({'Keuntungan':list_tax,'Total':tax_rate_transaction_num['Total Pajak Import']})

from scipy.stats import mannwhitneyu

# cek dengan uji statistika

H0 = 'Tidak ada perbedaan signifikan antara total pajak import pada negara profit dengan total pajak import pada negara rugi'
Ha = 'Terdapat perbedaan signifikan antara total pajak import pada negara profit dengan total pajak import pada negara rugi'
stat, pval = mannwhitneyu(df_uji[df_uji['Keuntungan']=='Profit']['Total'],
                          df_uji[df_uji['Keuntungan']=='Net Loss']['Total'])
if pval > 0.05:
    print(f"H0 diterima: {H0} (pvalue = {pval})")
else:
    print(f"H0 ditolak: {Ha} (pvalue = {pval})")
```

-   Berdasarkan hasil uji hipotesis dengan menggunakan Uji Mann-Whitney U, berikut adalah kesimpulan dari analisis yang dilakukan

-   Hasil Uji:
    -   Kesimpulan:
        -   Dengan p-value yang sangat rendah P-Value: 0.000179, ada cukup bukti untuk menolak hipotesis H0 dan menerima Ha. Ini menunjukkan bahwa terdapat perbedaan signifikan antara total pajak import pada negara profit dengan negara yang mengalami kerugian.

    -   Interpretasi:
        -   Ada Perbedaan Signifikan: Hasil ini mengindikasikan bahwa total pajak import di negara-negara yang memperoleh profit berbeda secara signifikan dari total pajak di negara-negara yang mengalami kerugian.

    -   Implikasi: 
        -   Perbedaan signifikan ini mengindikasikan bahwa pajak import di negara-negara yang mencatatkan profit sangat berbeda dengan negara-negara yang mengalami kerugian. Pajak import yang besar tampaknya berkontribusi terhadap kerugian, sementara pajak import yang lebih kecil cenderung mendukung profitabilitas. Oleh karena itu, diperlukan analisis lebih lanjut untuk memahami secara mendalam faktor-faktor yang menyebabkan perbedaan ini, sehingga perusahaan dapat merumuskan strategi yang lebih efektif.
```python
def rfm_score(baris):
    # Recency
    if baris['Recency'] <= 2:
        r = 4
    elif (baris['Recency'] > 3) and (baris['Recency'] <= 5):
        r = 3
    elif (baris['Recency'] > 6) and (baris['Recency'] <= 9):
        r = 2
    else:
        r = 1 # nilai paling kecil atau customer paling lama tidak membeli

    # Frequency
    if baris['Frequency'] <= 100:
        f = 1
    elif (baris['Frequency'] > 101) and (baris['Frequency'] <= 200):
        f = 2
    elif (baris['Frequency'] > 201) and (baris['Frequency'] <= 300):
        f = 3
    elif (baris['Frequency'] > 301) and (baris['Frequency'] <= 500):
        f = 4
    elif (baris['Frequency'] > 601) and (baris['Frequency'] <= 800):
        f = 5
    else:
        f = 6 # nilai paling besar atau customer yang sering membeli

    # Monetary:
    if baris['Monetary'] <= 500:
        m = 1
    elif (baris['Monetary'] > 501) and (baris['Monetary'] <= 700):
        m = 2
    elif (baris['Monetary'] > 701) and (baris['Monetary'] <= 900):
        m = 3
    else:
        m = 4 # nilai paling besar atau customer yang paling banyak rata-rata nominal belinya

    return r + f + m


rfm['RFM_Score'] = rfm.apply(rfm_score, axis = 1)
rfm.sort_values('RFM_Score',ascending=False).reset_index(drop=True)
```

```python
def rfm_segment(baris):
    if baris['RFM_Score'] > 10: # > 10
        return 'Champions'
    elif (baris['RFM_Score'] <= 10) and (baris['RFM_Score'] >= 8): # 8 - 10
        return 'Loyal Customers'
    elif (baris['RFM_Score'] < 8) and (baris['RFM_Score'] >= 5): # 5 - 8
        return 'Need Attantion'
    else:
        return 'Almost Lost / Hibernating' #  3-4

rfm['RFM_Segment'] = rfm.apply(rfm_segment, axis = 1)
rfm
```
